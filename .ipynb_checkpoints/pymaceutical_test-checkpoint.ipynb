## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
#Look at the mouse data
mouse_metadata
```

```python
# Look at the study results data
study_results
```

```python
# Combine the data into a single dataset using merge
Pymaceuts_df=pd.merge(mouse_metadata,study_results, on="Mouse ID", how="outer")

# Display the data table for preview
Pymaceuts_df.head()
```

```python
# Display the data table for review
Pymaceuts_df
```

```python
# counting unique values
MiceNumber = len(pd.unique(Pymaceuts_df["Mouse ID"]))
print("No of mice :",MiceNumber)
```

```python
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.nunique.html
# Count number of distinct elements in specified axis. Return Series with number of distinct elements
NumMice=Pymaceuts_df["Mouse ID"].count()
print("No of mice :",NumMice)
```

```python
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.nunique.html
# Count number of distinct elements in specified axis. Return Series with number of distinct elements
NumDupMice=Pymaceuts_df["Mouse ID"].nunique()
print("No of mice :",NumDupMice)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# https://pandas.pydata.org/docs/reference/api/pandas.Series.drop_duplicates.html?highlight=drop_duplicates
Clean_Pymaceuts=Pymaceuts_df.drop_duplicates(subset="Mouse ID", keep="last")
Clean_Pymaceuts.head()
```

```python
# Checking the number of mice in the clean DataFrame.
Clean_Pymaceuts.count()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use the groupby to splitting the object by Drug Regimen, applying a stats function, and combine the results
# This method is the most straighforward, creating multiple series and putting them all together at the end.
meanTumVol=Clean_Pymaceuts.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
medianTumVol=Clean_Pymaceuts.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
varTumVol=Clean_Pymaceuts.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
stdTumVol=Clean_Pymaceuts.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
semTumVol=Clean_Pymaceuts.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function
Sum_Stats_Table=pd.DataFrame({"Tumor volume mean":meanTumVol,
                             "Tumor volume median":medianTumVol,
                              "Tumor volume variance":varTumVol,
                              "Tumor volume standard deviation":stdTumVol,
                              "Tumor volume standard error of mean":semTumVol
                             })
# # Print on screen the Sum_Stats_Table 
Sum_Stats_Table.round(2)
```
