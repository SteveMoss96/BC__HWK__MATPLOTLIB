## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them all together at the end.

# Use the groupby to split the object by Drug Regimen, apply a stats function to tumor volume, and combine the results
meanTumVol=Clean_Pyma_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
medianTumVol=Clean_Pyma_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
varTumVol=Clean_Pyma_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
stdTumVol=Clean_Pyma_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
semTumVol=Clean_Pyma_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function

Sum_Stats_Table=pd.DataFrame({"Tumor volume mean":meanTumVol,
                             "Tumor volume median":medianTumVol,
                              "Tumor volume variance":varTumVol,
                              "Tumor volume standard deviation":stdTumVol,
                              "Tumor volume standard error of mean":semTumVol
                             })
# # Print on screen the Sum_Stats_Table 
Sum_Stats_Table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse - use .max to the find the greatest
Last_timepoint= Clean_Pyma_df.groupby('Mouse ID')['Timepoint'].max()
Last_timepoint_df=pd.DataFrame(Last_timepoint)
# Last_timepoint

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
Merged=pd.merge(Clean_Pyma_df,Last_timepoint_df, on=("Mouse ID", "Timepoint"), how="right")
# Merged
```

```python
# Merged.info()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
Drugs_list=['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
# Create empty list to fill with tumor vol data (for plotting)
Tumor_Vol=[]

# Locate each drug from our list 'drugs' in the orginal dataframe
for value in Drugs_list:
    Top4Mice=Merged.loc[Merged["Drug Regimen"]==value,"Tumor Volume (mm3)"]
    Tumor_Vol.append(Top4Mice)    

    # Calculate the IQR and quantitatively determine if there are any potential outliers. 
    quartiles=Top4Mice.quantile([0.25, 0.5, 0.75])
    lowerq = quartiles[0.25]
    upperq=quartiles[0.75]
    iqr=upperq-lowerq
    upper_bound=upperq+(1.5*iqr)
    lower_bound=lowerq-(1.5*iqr)
#     print(f"Values below {lower_bound} could be outliers.")
#     print(f"Values above {upper_bound} could be outliers.")
    outliers=Top4Mice.loc[(Top4Mice<lower_bound) | (Top4Mice> upper_bound)]
    print(f"{value} {outliers}")
#     print(outliers)
#     print(Top4Mice)
# Locate the rows which contain mice on each drug and get the tumor volumes  
 
# Determine outliers using upper and lower bounds
# ????how do I create a subset dataframe only with rows only containing the final timepoint for 4 selected regimens    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.boxplot(Tumor_Vol, labels=Drugs_list)
plt.xlabel("Drug Regimen")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Final tumor volume for selected regimens, Boxplot")
plt.show()
```

```python
# https://matplotlib.org/stable/gallery/statistics/boxplot.html?highlight=flierprops
# https://towardsdatascience.com/create-and-customize-boxplots-with-pythons-matplotlib-to-get-lots-of-insights-from-your-data-d561c9883643
# flierprops: dictionary of properties applied to outliers.
# medianprops: dictionary of properties applied to median line inside the boxplot.
# whiskerprops: dictionary of properties applied to the whiskers.
# capprops: dictionary of properties applied to the caps on the whiskers.
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
flierprops = dict(marker='D', markerfacecolor='pink', markersize=15, markeredgecolor='red')
plt.boxplot(Tumor_Vol, labels=Drugs_list,
            notch=True, 
            vert=True, 
            patch_artist=True,
            flierprops=flierprops)
plt.xlabel("Drug Regimen")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Final tumor volume for selected regimens, Boxplot")
plt.show()    
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen  ((f"and p={p_val:.15f}")))
r_val, p_val = st.pearsonr(avg_vol["Weight (g)"], avg_vol["Tumor Volume (mm3)"])
print(f"The r correlation coefficient of mouse wt and av tumor vol for Capomulin regimen is r={r_val:.3f}")
```

```python
# # Create equation of line to calculate predicted violent crime rate
# vc_fit = vc_slope * year + vc_int

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(avg_vol["Weight (g)"], avg_vol["Tumor Volume (mm3)"])
regression = slope * avg_vol + intercept
line_eq = "y = " + str(round(slope,4)) + "x + " + str(round(intercept,4))
plt.scatter(avg_vol["Weight (g)"], avg_vol["Tumor Volume (mm3)"], marker = 'D', color="crimson")
plt.plot(avg_vol, regression, "--", color="orange")
plt.title("Av Mouse Weight versus Av Tumor Volume  with regression, Capomulin Regimen")
plt.xlabel("Average Mouse Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.ylim([33, 47])
plt.xlim([14.5, 25.25])

# Annotate linear regression
plt.annotate(line_eq,(19,35),fontsize=10,color="purple")
plt.show()
```
